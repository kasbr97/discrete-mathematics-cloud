<h4>Subset</h4>
<p>A set X is a subset of set Y (Written as X  Y) if every element of X is an element of set Y. For example:</p>
<ul>
    <li>Given two sets, X={1,2,3,4,5,6} and Y={1,2}. Y is a subset of set X as all the elements of set Y are in set X. Hence, we can write Y  X.</li>
    <li>Given two sets, X={1,2,3} and Y={1,2,3}. Y is a subset of set X as all the elements of set Y are in set X. Hence, we can write Y  X.</li>
</ul>

<p>The previous example can be programmed in Python to obtain a quick answer. Using the method "issubset()", which will return a boolean to indicate if the set is a subset of another.</p>

```python
X = {1, 2, 3, 4, 5, 6}
Y = {1, 2, 3, 4, 5, 6}
Y.issubset(X)
```

<h4>Proper Subset</h4>
<p>The term proper subset can be defined as subset of but not equal to. A Set X is a proper subset of set Y (Written as X  Y) if every element of X is an element of set Y and |X|<|Y|. For example: </p>
    <ul>
        <li>Given two sets, X={1,2,3,4,5,6} and Y={1,2}. Here set YX since all elements in Y are contained in X too and X has at least one element is more than set Y.</li>
  </ul>
    
<p>The previous example can be programmed in Python to obtain a quick answer. Using the "<" operator, which will return a boolean to indicate if the set is a proper subset of another.</p>

```python
X = {1, 2, 3, 4, 5, 6}
Y = {1, 2, 3, 6}
Y < X
```

<h2>Set Operations</h2>
<p>Set Operations include Set Union, Set Intersection, Set Difference, Complement of Set, and Cartesian Product.</p>
<h4>Set Union</h4>
<p>The union of sets A and B (denoted by AB) is the set of elements which are in A, in B, or in both A and B. Hence, AB={x|xA OR xB}.</p>
<ul>
    <li>If A={10,11,12,13} and B = {13,14,15}, then AB={10,11,12,13,14,15}. (The common element occurs only once). The following Venn Diagram ilustrates the result from the set operation.</li>
</ul>



<p>In Python, the method "union()" is used for this operation.</p>

```python
A = {10, 11, 12, 13}
B = {13, 14 ,15}
A.union(B)
```

<h4>Set Intersection</h4>
<p>The intersection of sets A and B (denoted by A  B) is the set of elements which are in both A and B. Hence, A  B={x|xA AND xB}.</p>
<ul>
    <li>Given the sets A={11,12,13} and B={13,14,15}, then AB={13}. The following Venn Diagram ilustrates the result from the set operation.</li>
<ul>



<p>The method used to represent the operation in Python is "intersection()".</p>

```python
A = {11, 12, 13}
B = {13, 14, 15}
A.intersection(B)
```

<h4>Set Difference / Relative Complement</h4>
<p>The set difference of sets A and B (denoted by AB) is the set of elements which are only in A but not in B. Hence, AB={x|xA AND xB}.</p>
<ul> 
    <li>A={10,11,12,13}  and B={13,14,15}, then (AB)={10,11,12} and (BA)={14,15}. Here, we can see (AB)(BA). The following Venn Diagram ilustrates the result from the set operation. A-B (left), B-A (right).</li>
</ul>
    


<p>The method used to represent a set difference in Python is "difference()". The following example creates two sets A and B, and prints the result for the difference between A-B and B-A.</p>

```python
A = {10, 11, 12, 13}
B = {13, 14, 15}

#A-B
print(A.difference(B))

#B-A
print(B.difference(A))
```

<h4>Complement of a Set</h4>
<p>The complement of a set A (denoted by A) is the set of elements which are not in set A. Hence, A={x|xA}. More specifically, A=(UA) where U is a universal set which contains all objects.</p>
<ul>
    <li>If the set A={ x|x belongs to set of odd integers }  then A={ y|y does not belong to set of odd integers } The Venn Diagram shows the result of A highlighted in yellow.</li>
</ul>
    


<p>Given a universal set U = {1, 5, 8, 2, 4, 3, 9} and a set A = {8, 5, 3, 9}, the complement of set A can be represented in Python in by the following</p>

```python
U = {1, 5, 8, 2, 4, 3, 9}
A = {8, 5, 3, 9}

print("A = ", U.difference(A))
```

<h4>Cartesian Product / Cross Product</h4>
<p>The Cartesian product of n number of sets A1,A2,An denoted as A1A2An can be defined as all possible ordered pairs (x1,x2,xn) where x1A1,x2A2,xnAn</p>
<p> For example, if given the two sets A={a,b} and B={1,2}, </p>
<ul>
     <li>The Cartesian product of A and B is written as AB={(a,1),(a,2),(b,1),(b,2)}. </li>
     <li>The Cartesian product of B and A is written as BA={(1,a),(1,b),(2,a),(2,b)}.</li>
</ul>
<p>To replicate the previous example in Python, two sets are created then a list that contains lists (to keep an order). The list cartesian_productAB uses two for-loops to run through both sets:</p>

```python
A = {'a', 'b'}
B = {1, 2}

#A x B
cartesian_productAB = [ [Ax, Bx] for Ax in A for Bx in B]

#B x A
cartesian_productBA = [ [Bx, Ax] for Bx in B for Ax in A]


print(cartesian_productAB)
print(cartesian_productBA)
```

<h2>Power Set</h2>
<p>Power set of a set S is the set of all subsets of S including the empty set. The cardinality of a power set of a set S of cardinality n is 2^n. Power set is denoted as P(S).</p>

<p>For a set S={a,b,c,d} let us calculate the subsets</p>
<ul>
    <li>Subsets with 0 elements  {} (the empty set)</li>
    <li>Subsets with 1 element  {a},{b},{c},{d} </li>
    <li>Subsets with 2 elements  {a,b},{a,c},{a,d},{b,c},{b,d},{c,d}</li>
    <li>Subsets with 3 elements  {a,b,c},{a,b,d},{a,c,d},{b,c,d}</li>
    <li>Subsets with 4 elements   {a,b,c,d}</li>
</ul>

<p>Hence, P(S)= { {},{a},{b},{c},{d},{a,b},{a,c},{a,d},{b,c},{b,d},{c,d},{a,b,c},{a,b,d},{a,c,d},{b,c,d},{a,b,c,d} }</p>
<p>|P(S)|= 2^4 = 16</p>

<p>To replicate the previous example in Python, it is needed to use the function combinations from itertools. The function combinations() takes a list and an integer as parameters. In the code, the integer is "i", which starts at cero, therefore the function makes different combinations of the items in the list (s) as many times as the number "i" indicates it. </p>

```python
from itertools import combinations
def print_powerset(s, new_s):
    for i in range(0,len(s)+1):
        for element in combinations(s,i):
            #print(''.join(element))
            new_s.append(element)
    print(new_s)
s=['a','b','c']
new_s = []
print_powerset(s, new_s)

```

<h2>Partitioning of a Set</h2>
<p>Partition of a set, say S, is a collection of n disjoint subsets, say P1,P2,Pn that satisfies the following three conditions</p>
<ol>
    <li>Pi does not contain the empty set. </li>
    <li>The union of the subsets must equal the entire original set.</li>
    <li>The intersection of any two distinct sets is empty.</li>
</ol>

<p>An example of partitioning of a set would be the following: Given a set S={a,b,c,d,e,f,g,h}, one probable partitioning is {a},{b,c,d},{e,f,g,h}. Another probable partitioning is {a,b},{c,d},{e,f,g,h}</p>

<h4>Bell Numbers</h4>
<p>Bell numbers give the count of the number of ways to partition a set. They are denoted by Bn where n is the cardinality of the set.</p>
<p>Example: Given the set S={1,2,3}, n=|S|=3. The alternate partitions are:</p>
<ol>
<li>,{1,2,3}</li>

<li>{1},{2,3}</li>

<li>{1,2},{3}</li>

<li>{1,3},{2}</li>

<li>{1},{2},{3}</li>
</ol>
<p>Hence B3=5</p>
